```python
# Function to extract the last number
def get_last_number(input_string):
    current_number = ""
    last_number = None

    for char in input_string:
        if char.isdigit():
            current_number += char
        else:
            if current_number:
                last_number = current_number
                current_number = ""

    # Check for trailing number
    if current_number:
        last_number = current_number

    return int(last_number) if last_number else None
```

```python
# Calculate frequencies and standard errors for first messages
first_message_counts = (
    df[(df['sequence_number'] == 1) & (df['role'] == 'assistant')]
    .groupby(['arm'])['rating']
    .value_counts()
    .unstack(fill_value=0)
)

first_message_counts.columns = ['first_message_count_-1', 'first_message_count_0', 'first_message_count_1']

# Total counts
first_message_counts['first_message_count'] = first_message_counts[['first_message_count_-1', 'first_message_count_0', 'first_message_count_1']].sum(axis=1)

# Frequencies
first_message_counts['first_message_freq_-1'] = first_message_counts['first_message_count_-1'] / first_message_counts['first_message_count']
first_message_counts['first_message_freq_0'] = first_message_counts['first_message_count_0'] / first_message_counts['first_message_count']
first_message_counts['first_message_freq_1'] = first_message_counts['first_message_count_1'] / first_message_counts['first_message_count']

# Standard errors
first_message_counts['first_message_freq_-1_se'] = np.sqrt(first_message_counts['first_message_freq_-1'] * (1 - first_message_counts['first_message_freq_-1']) / first_message_counts['first_message_count'])
first_message_counts['first_message_freq_0_se'] = np.sqrt(first_message_counts['first_message_freq_0'] * (1 - first_message_counts['first_message_freq_0']) / first_message_counts['first_message_count'])
first_message_counts['first_message_freq_1_se'] = np.sqrt(first_message_counts['first_message_freq_1'] * (1 - first_message_counts['first_message_freq_1']) / first_message_counts['first_message_count'])

first_message_counts = first_message_counts.reset_index()

# Calculate frequencies and standard errors for follow-up messages
follow_up_counts = (
    df[(df['sequence_number'] != 1) & (df['role'] == 'assistant')]
    .groupby(['arm'])['rating']
    .value_counts()
    .unstack(fill_value=0)
)

follow_up_counts.columns = ['follow_up_count_-1', 'follow_up_count_0', 'follow_up_count_1']

follow_up_counts['follow_up_count'] = follow_up_counts[['follow_up_count_-1', 'follow_up_count_0', 'follow_up_count_1']].sum(axis=1)

follow_up_counts['follow_up_freq_-1'] = follow_up_counts['follow_up_count_-1'] / follow_up_counts['follow_up_count']
follow_up_counts['follow_up_freq_0'] = follow_up_counts['follow_up_count_0'] / follow_up_counts['follow_up_count']
follow_up_counts['follow_up_freq_1'] = follow_up_counts['follow_up_count_1'] / follow_up_counts['follow_up_count']

# Standard errors
follow_up_counts['follow_up_freq_-1_se'] = np.sqrt(follow_up_counts['follow_up_freq_-1'] * (1 - follow_up_counts['follow_up_freq_-1']) / follow_up_counts['follow_up_count'])
follow_up_counts['follow_up_freq_0_se'] = np.sqrt(follow_up_counts['follow_up_freq_0'] * (1 - follow_up_counts['follow_up_freq_0']) / follow_up_counts['follow_up_count'])
follow_up_counts['follow_up_freq_1_se'] = np.sqrt(follow_up_counts['follow_up_freq_1'] * (1 - follow_up_counts['follow_up_freq_1']) / follow_up_counts['follow_up_count'])

follow_up_counts = follow_up_counts.reset_index()

# Merge the two DataFrames on 'arm'
merged_counts = pd.merge(first_message_counts, follow_up_counts, on='arm', how='outer').fillna(0)

# Compute conversation lengths per chat_id and arm
conversation_lengths = (
    df[df['role'] == 'assistant']
    .groupby(['chat_id', 'arm'])
    .size()
    .reset_index(name='conversation_length')
)

# Calculate statistics per arm
conversation_stats = conversation_lengths.groupby('arm')['conversation_length'].agg(['mean', 'std', 'count']).reset_index()

# Standard error and 95% confidence interval
conversation_stats['se'] = conversation_stats['std'] / np.sqrt(conversation_stats['count'])

# Merge conversation stats with merged_counts
merged_counts = merged_counts.merge(conversation_stats[['arm', 'mean', 'se']], on='arm', how='left')

# Rename columns for clarity
merged_counts.rename(columns={'mean': 'average_conversation_length',
                              'se': 'average_conversation_length_se'}, 
                     inplace=True)



differential_messages['diagnosis_present'] = differential_messages["evaluation_count"].astype(bool)
sens = pd.DataFrame(differential_messages.groupby("arm")['diagnosis_present'].mean())
sens.columns = ['sensitivity']
#sens['sensitivity_se'] = np.sqrt(sens['sensitivity'] * (1 - sens['sensitivity']) / sens['sensitivity'])
sens['sensitivity_se'] = differential_messages.groupby("arm")['diagnosis_present'].std() / np.sqrt(differential_messages.groupby("arm")['diagnosis_present'].count())

sens = sens.reset_index()
display(sens)

differential_messages["differentials_length"] = differential_messages["differentials"].apply(json.loads).apply(len)
differential_messages["average_correct"] = differential_messages["evaluation_count"] / differential_messages["differentials_length"]
spec = pd.DataFrame(differential_messages.groupby("arm")["average_correct"].mean())
spec.columns = ['specificity']
spec['specificity_se'] = differential_messages.groupby("arm")["average_correct"].std() / np.sqrt(differential_messages.groupby("arm")["average_correct"].count())

spec = spec.reset_index()
display(spec)

merged_counts = merged_counts.merge(sens, on="arm").merge(spec, on="arm")

merged_counts.T
```

```python
def plot_diverging_bars(ax, data, freq_up_col, freq_down_col, title, bar_colors=None, show_error_bars=False, freq_up_se_col=None, freq_down_se_col=None):
    arms = data['arm']
    x_positions = np.arange(len(arms))
    
    # For each position, we'll plot with its own color
    ax.axhline(0, color='black', linewidth=2, linestyle='--')  # Baseline at y=0
    for i, x_pos in enumerate(x_positions):
        # Upvote bars with optional error bars
        up_kwargs = {
            'color': bar_colors['up_left'] if i == 0 else bar_colors['up_right'],
            'edgecolor': 'black',
            'linewidth': 2
        }
        if show_error_bars and freq_up_se_col:
            up_kwargs['yerr'] = data[freq_up_se_col].iloc[i]
            up_kwargs['capsize'] = 5
        
        # Downvote bars with optional error bars
        down_kwargs = {
            'color': bar_colors['down_left'] if i == 0 else bar_colors['down_right'],
            'edgecolor': 'black',
            'linewidth': 2
        }
        if show_error_bars and freq_down_se_col:
            down_kwargs['yerr'] = data[freq_down_se_col].iloc[i]
            down_kwargs['capsize'] = 5
        
        # Plot bars individually
        ax.bar(
            x_pos, 
            data[freq_up_col].iloc[i], 
            label='Upvote' if i == 0 else None,  # Only label once
            width=0.8,
            **up_kwargs
        )
        
        ax.bar(
            x_pos, 
            -data[freq_down_col].iloc[i], 
            label='Downvote' if i == 0 else None,  # Only label once
            width=0.8,
            **down_kwargs
        )

    # Rest of formatting remains the same...
    for i in [0, 1]:
        ax.plot([i-0.4, i+0.4], 
                [data[freq_up_col][i]-data[freq_down_col][i],
                 data[freq_up_col][i]-data[freq_down_col][i]],
                '-',
                linewidth=2,
                color='white',
                label='Difference' if i == 0 else None,
               )
    
    # Adding numerical values above/below bars
    for i in range(len(x_positions)):
        up_value = data[freq_up_col].iloc[i]
        down_value = data[freq_down_col].iloc[i]
        ax.text(i+0.016, up_value + 0.11, f"{up_value * 100:.2f}%", ha='center', fontsize=16, fontweight='bold')
        ax.text(i+0.016, -down_value - 0.16, f"{down_value * 100:.2f}%", ha='center', fontsize=16, fontweight='bold')

    # Assuming 'arm' is the categorical variable on the y-axis
    data['difference'] = data[freq_up_col] - data[freq_down_col]
    categories = data['arm']
    # Loop over each category to add the difference value
    for idx, category in enumerate(categories):
        row = data[data['arm'] == category].iloc[0]
        diff_value = row['difference']
        freq_up = row[freq_up_col]
        freq_down = row[freq_down_col]

        # Determine x position of the white line (midpoint between bars)
        x_pos = idx

        # Determine y position
        y_pos = diff_value  # or adjust based on how y-ticks are set

        y_offset = -0.08  # Adjust as needed

        # Format the label as per other numerical labels
        label = ("+" if diff_value > 0 else "-") + f"{np.abs(diff_value)*100:.2f}%"

        # Add text annotation
        ax.text(x_pos, y_pos + y_offset, label, ha='center', fontsize=16, fontweight='bold', color='white')
    
    # Formatting

    ax.set_xticks(x_positions)
    ax.set_xticklabels(arms, fontsize=16, fontweight='bold')
    ax.set_ylim(-1, 1)  # Set consistent y-axis range
    ax.set_yticks(np.linspace(-1, 1, 9),
                  np.round(np.linspace(-1, 1, 9), 2),
                  fontsize=16,
                  fontweight='bold')
    ax.set_title(title, fontsize=18, fontweight='bold', pad=10)
    ax.set_ylabel('Frequency', fontsize=16, fontweight='bold')
    ax.set_xlim(-0.5, 1.5)
    
    # Thicker ticks and spines
    ax.tick_params(axis='both', which='both', width=2)
    ax.spines['top'].set_linewidth(2)
    ax.spines['right'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)
    # ax.legend(fontsize=12)

def plot_general_bars(ax, data, values_col, title, left_bar_color, right_bar_color, y_label='Value', comparison_line=None, percentage=True, show_error_bars=False, se_col=None, label_offset=0.04, ylim_offset=0.1, yticks=1):
    arms = data['arm']
    x_positions = np.arange(len(arms))
    
    # Iterate over the positions to plot bars with alternating colors
    for i, x_pos in enumerate(x_positions):
        bar_kwargs = {
            'color': left_bar_color if i == 0 else right_bar_color,
            'edgecolor': 'black',
            'linewidth': 2
        }
        if show_error_bars and se_col:
            bar_kwargs['yerr'] = data[se_col][i]
            bar_kwargs['capsize'] = 5
        
        ax.bar(x_pos, data[values_col][i], **bar_kwargs)
        
        # Add numerical values above bars
        value = data[values_col][i]
        formatted_value = f"{value * 100:.2f}%" if percentage else f"{value:.2f}"
        ax.text(x_pos+0.016, value + label_offset, formatted_value, ha='center', fontsize=16, fontweight='bold')
    
    # Optional comparison line
    if comparison_line is not None:
        ax.plot([-0.5, len(arms)-0.5], [comparison_line, comparison_line], '--', color='k', linewidth=2)
    
    # Formatting
    ax.set_xticks(x_positions)
    ax.set_xticklabels(arms, fontsize=16, fontweight='bold')
    ax.set_title(title, fontsize=18, fontweight='bold', pad=10)
    ax.set_ylabel(y_label, fontsize=16, fontweight='bold')
    
    # Set consistent ylim based on ylim_offset
    if percentage:
        max_value = 1.00
        ax.set_yticks(np.linspace(0, 1, 5),
                      np.round(np.linspace(0, 1, 5), 2),
                      fontsize=16,
                      fontweight='bold')
    else:
        max_value = np.ceil(max(data[values_col])) + ylim_offset
        ax.set_yticks(np.linspace(0, max_value, yticks).astype(int),
                      np.linspace(0, max_value, yticks).astype(int),
                      fontsize=16,
                      fontweight='bold')
    ax.set_ylim(0, max_value)
    ax.set_xlim(-0.5, len(arms)-0.5)
    
    # Thicker ticks and spines
    ax.tick_params(axis='both', which='both', width=2)
    for spine in ['top', 'right', 'left', 'bottom']:
        ax.spines[spine].set_linewidth(2)
```
